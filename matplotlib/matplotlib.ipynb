## Lecture 10 Subplots 

```python

import pandas as pd
from matplotlib import pyplot as plt


# Reading and accessing the data
data = pd.read_csv('/Users/ironfeast/Desktop/python_project/matplotlib/data5.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

# This one helps to plot the graph on the multiple axis  for the given data 

fig1, ax1 = plt.subplots()
fig2, ax2 = plt.subplots()

ax1.plot(ages, dev_salaries, color='#444444',
         linestyle='-.', label='All Devs', linewidth=2)

ax2.plot(ages, py_salaries, label='Python',linewidth=2)
ax2.plot(ages, js_salaries, label='JavaScript',linewidth=2)

ax1.legend(loc='best')
ax1.set_title('Median Salary (USD) by Age')
ax1.set_ylabel('Median Salary (USD)')

ax2.legend()
ax2.set_xlabel('Ages')
ax2.set_ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()

fig1.savefig('fig1.png')
fig2.savefig('fig2.png')
```
