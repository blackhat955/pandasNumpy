## Lecture 9 Cleaning and clustering 

```python
import pandas as pd
import numpy as np
people = {
    'first': ['Corey', 'Jane', 'John', 'Chris', np.nan, None, 'NA'], 
    'last': ['Schafer', 'Doe', 'Doe', 'Schafer', np.nan, np.nan, 'Missing'], 
    'email': ['CoreyMSchafer@gmail.com', 'JaneDoe@email.com', 'JohnDoe@email.com', None, np.nan, 'Anonymous@email.com', 'NA'],
    'age': ['33', '55', '63', '36', None, None, 'Missing']
}
```

```python
handle = pd.DataFrame(people)
```

```python
handle
```

```python
# drop NAN in given data frame
handle.dropna()
```

```python
# now put default argument to delete data
# how-> can be aby or all any for delete if some of them are null and all if all of them null
handle.dropna(axis='index', how='all')
```

```python
# This one deletes rows where email is not there and other elements represent if this much is present in the table then the rows is not going to delete 
# all mean that there is drop of the row if there is no attribute present which is mentioned in subset 
handle.dropna(axis='index', how='all',subset=['email', 'last'])
```

```python
# replace other NA and missing with np.nan to better handle it 
handle.replace('NA',np.nan, inplace=True)
handle.replace('Missing',np.nan, inplace=True)
```

```python
handle
```

```python
handle.dropna()
```

```python
# to know about attribute type
handle.dtypes
```

```python
handle['age']=handle['age'].astype(float)
```

```python
handle['age'].mean()
```
